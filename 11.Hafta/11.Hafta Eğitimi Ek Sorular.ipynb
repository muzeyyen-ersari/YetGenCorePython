# Problem 3
Her bir sutuna ait unique değerlerin sayısını bulunuz.

```python
for column in titanic.columns:
    unique_values = titanic[column].nunique()
    print(f"{column}: {unique_values}")
```

# Problem 7
embarked değeri C olanların tüm bilgilerini gösteriniz.

```python
import pandas as pd
embarked_c = []

for i in range(len(titanic)):
    if titanic.loc[i, "embarked"] == "C":
        embarked_c.append(titanic.loc[i,:])

embarked_c = pd.concat(embarked_c, axis=1).T

print(embarked_c)
   
```

```python
embarked_c = titanic.loc[titanic["embarked"] == "C"]

print(embarked_c)
```

# Problem 8
